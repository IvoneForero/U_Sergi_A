```python
# Suna y resta DE MATRICES
import numpy as np

a = np.array([[1,2,3],[4,5,6]])
b = np.array([[7,8,3],[1,3,2]])

print('Suma')
print(a + b)
print('Resta')
print(a - b)

print('Suma')
print(np.add(a, b))

print('Resta')
print(np.subtract(a, b))

print('Raiz cuadrada')
print(np.sqrt(a))

print('Potencia')
print(np.power(a, 2))

#La función sign(array [, out]) se utiliza para indicar el signo de un elemento numérico. Para entradas enteras, si el valor de la matriz es mayor que 0, 
#devuelve 1, si el valor de la matriz es menor que 0, devuelve -1, y si el valor de la matriz es 0, devuelve 0

print('Signo')
print(np.sign(a))

```

```python
#Los gráficos pueden mostrar texto en los ejes porque estos se basan en valores numéricos.
# %%
import numpy as np
import matplotlib.pyplot as plt

ventas = np.random.randint(150, size=[6])
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio']
mapeado = range(len(meses))

plt.plot(ventas)           # se agrega el gráfico
plt.xticks(mapeado, meses)  # se mapean los valores horizontales
plt.show()                  # visualización
```

```python
# %%
# Mostramos las ventas de tres personas diferentes
import numpy as np
import matplotlib.pyplot as plt

meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio']
mapeado = range(len(meses))

plt.plot(np.random.randint(100, size=[6]), label="Rodrigo")
plt.plot(np.random.randint(100, size=[6]), label="Miriam")
plt.plot(np.random.randint(100, size=[6]), label="Erika")
#plt.xlim(2, 4)")Z"
plt.title("Ventas del primer semestre")
plt.xlabel("Meses")
plt.ylabel("Cantidad en $")
plt.legend()
plt.show()
```

```python
# %%
# Personalizamos las líneas
import numpy as np
import matplotlib.pyplot as plt

meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio']
mapeado = range(len(meses))

plt.plot(np.random.randint(100, size=[6]),
         label="Rodrigo", color="purple", ls="-", lw="2")
plt.plot(np.random.randint(100, size=[6]),
         label="Miriam", color="#9811ff", ls="--", lw="2")
plt.plot(np.random.randint(100, size=[6]),
         label="Alexander", color="orange", ls="-.", lw="2")
plt.xticks(mapeado, meses)
plt.legend()
plt.show()
```

```python
# %%
# Personalizamos los marcadores

meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio']
mapeado = range(len(meses))

plt.plot(np.random.randint(100, size=[6]),
         marker="o", markersize="8", markeredgewidth="2",
         markerfacecolor="green", markeredgecolor="white")
plt.plot(np.random.randint(100, size=[6]),
         marker="*", markersize="10", markeredgewidth="2",
         markerfacecolor="red", markeredgecolor="white")
plt.plot(np.random.randint(100, size=[6]),
         marker="D", markersize="5", markeredgewidth="2",
         markerfacecolor="orange", markeredgecolor="white")
plt.xticks(mapeado, meses)
plt.legend()
plt.show()
```

```python
# %%
# Dibujos con subgráficos
import numpy as np
import matplotlib.pyplot as plt

plt.subplot(1, 3, 1)  # Tabla 1x3 y dibujaremos en la celda 1
plt.plot(np.random.randint(100, size=[6]), label="Rodrigo", color="green")
plt.ylim(0, 100)
plt.legend()

plt.subplot(1, 3, 2)  # Tabla 1x3 y dibujaremos en la celda 2
plt.plot(np.random.randint(100, size=[6]), label="Alexander", color="red")
plt.ylim(0, 100)
plt.legend()

plt.subplot(1, 3, 3)  # Tabla 1x3 y dibujaremos en la celda 3
plt.plot(np.random.randint(100, size=[6]), label="Adriana", color="cyan")
plt.ylim(0, 100)
plt.legend()

plt.show()  # Dibujamos el conjunto
```

```python
# %%
# Dibujando 9 subgráficos
for i in range(9):
    plt.subplot(3, 3, i+1)  # Tabla 3x3
    plt.plot(np.random.randint(100, size=[6]))
    #plt.plot(np.random.randint(100, size=[6]))
    plt.ylim(0, 100)

plt.show()
```
